```python
# Comparative Analysis of Random Forest and ANN  
## Network Intrusion Detection using CICIDS Dataset

---

## üéØ Project Objective
To compare the performance of:
- Random Forest (Machine Learning)
- Artificial Neural Network (Deep Learning)

Under:
- Binary Classification
- Multiclass Classification

---

## üìä Dataset
- CICIDS Network Traffic Dataset
- 78 Features
- Stratified 70-30 Train-Test Split

---

## üî¨ Binary Classification Results

| Model | Accuracy | Macro F1 |
|--------|----------|----------|
| Random Forest | 0.9989 | 0.9983 |
| ANN | 0.9883 | 0.9816 |

### Observation:
Random Forest slightly outperformed ANN in binary classification.

---

## üî¨ Multiclass Classification Results

| Model | Accuracy | Macro F1 |
|--------|----------|----------|
| Random Forest | 0.9985 | 0.9125 |
| ANN Regular | 0.9850 | 0.7763 |
| ANN Focal | 0.9822 | 0.7740 |
| ANN SMOTE | 0.9903 | 0.7379 |
| ANN ClassWeight | 0.8990 | 0.5823 |

### Observation:
- Random Forest clearly performs best overall.
- ANN struggles with minority classes.
- Class weighting reduced overall stability.
- SMOTE improved accuracy but not Macro F1.

---

## üèÜ Final Conclusion

Random Forest is the most reliable model for this dataset in both binary and multiclass settings.

ANN requires further architectural tuning to match ensemble-based ML models.
```
